```python
#sf = ['hello world\n','today is Saturday\n', 'winter is comming\n']
sf5pots = []
pots_side = 72
pot = 901
repeat = 7
for p in range(4):
    if p>=1:
        pot = pot + 100
    for i in range(pot,pot+pots_side):
        #print(f'{i:04d}\\n')
        for l in range(repeat):
            sf5pots.append(f'{i:04d}\n')
            l+=1

print(sf5pots)
with open('pots5.txt', 'w') as f:  # abre um arquivo para escrita chamado 'data2write.txt'
      f.writelines(sf5pots)       # escreve o conteúdo da lista 'lst' no arquivo
```
