### Exercício 5)
Qual foi o estádio com o maior número de partidas?

a) Nacional<br>
b) Jalisco<br>
c) Maracana<br>
d) Estadio Azteca

__Dica:__ Construa um dicionário onde a chave é o nome do estádio e o valor do número de partidas disputadas naquele estádio. Busque pela tupla (chave,valor) onde valor é máximo.

````python
stadiummatches = {}

for i in range(copa_partidas.shape[0]):
    stadium = copa_partidas.iloc[i]['Stadium']
    if stadium not in stadiummatches:
        stadiummatches[stadium] = 1
    else:
        stadiummatches[stadium] += 1

max_mtc = 0

for k,v in stadiummatches.items():
    if v > max_mtc:
        max_st = k
        max_mtc = v

print(f"Estádio com maior número de partidas: {max_st} - {max_mtc} jogos")
````
