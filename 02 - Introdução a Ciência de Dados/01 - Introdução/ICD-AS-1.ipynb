# (1) Determine a moda do conjunto:

```python
Z = [0,0,2,3,1,2,4,5,6,8,0,0,7,0,1,2,1,1,5,3,1,0,7,3,2,3,4,5,6,7,8,9,1,1,2,2,3]

Escolha uma opção:
a. 1
b. 2
c. 5
d. 3
e. 7
```

````python
import numpy as np 
from statistics import mode
import matplotlib.pyplot as plt

Z = [0,0,2,3,1,2,4,5,6,8,0,0,7,0,1,2,1,1,5,3,1,0,7,3,2,3,4,5,6,7,8,9,1,1,2,2,3]

print('Z:', Z)
print('Order(Z): ', np.sort(Z) )
print('Unique(Z): ', np.unique(Z) )

freq = np.zeros(len(np.unique(Z))) # armazena as frequencias
ind = 0 # indice do vetor de frequências
for i in np.unique(Z): # para os valores diferentes
    counter = 0 # conta as ocorrências
    for j in range(0,len(Z)):
        if(Z[j] == i):
            counter = counter + 1
    freq[ind] = counter
    ind = ind + 1
print('Freq: ', freq )
m = mode(Z)
print('A moda de Z:', m)
````

# (2) Determine a média e mediana de Z


````python
muz = np.mean(Z)
print('Mean(Z):', muz)

mdz = np.median(Z)
print('Median(Z):', mdz)

y_pos = np.arange(len(np.unique(Z)))
plt.xticks(y_pos, np.unique(Z), label="data")
plt.bar(y_pos, freq)
plt.xlabel("Valores",fontsize = 20) 
plt.ylabel("Frequência", fontsize = 20) 

plt.vlines(muz,0,max(Z), color='red', label="mean")
plt.vlines(mdz,0,max(Z), linestyles = 'dashed', color='red', label="median")
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("x",fontsize = 20) 
plt.ylabel("p(x)", fontsize = 20) 

plt.legend()
plt.show()
````

# (3) Calcule a variância, o desvio padrão e distância interquantil de Z':
```python
Z' = [0,0,2,3,1,2,4,5,6,8,7,5,3,1,0,7,3,2,3,4,5,6,7,8,9,1,1,2,2,3]
```

````python
from scipy.stats import iqr

def variancia(X):
    m = np.mean(X)
    N = len(X)
    s = 0
    for i in np.arange(0, len(X)):
        s = s + (X[i]-m)**2
    s = s/(N-1)
    return s

Z_ = [0,0,2,3,1,2,4,5,6,8,7,5,3,1,0,7,3,2,3,4,5,6,7,8,9,1,1,2,2,3]
print('Z_:', Z_)
print('Order(Z_): ', np.sort(Z_) )
print('Unique(Z_): ', np.unique(Z_) )

print('Variância de Z_:', variancia(Z_))
print('Desvio Padrão de Z_:', np.sqrt(variancia(Z_)))
print('IQR de Z_:', iqr(Z_))

import pandas as pnd

zdf = pnd.DataFrame(Z_)
zdf.describe()
````

# (4) Para a função tangente no intervalo `[-1.5,1.5]`, calcule os coeficientes de Pearson e Spearman.
Escolha uma opção:
```python
a. 1 e 1
b. 0.8 e 0.7
c. 0.99 e 0.5
d. 0.8 e 0.99
e. 0.8 e 0.6
```

````python
from scipy.stats import pearsonr, spearmanr

N = 100
X = np.linspace(-1.5,1.5, N)
#erro = np.random.uniform(-1.5,1.5,N) # ruído a ser incluído na relação linear.
#erro = np.random.uniform(-1.5,1.5,N) # ruído a ser incluído na relação linear.
Y = np.tan(X)
plt.plot(X,Y, 'ro')
corr, p_value = pearsonr(X, Y) # calcula a correlação
corr = int(corr*100)/100
plt.xlim(-2,2)
plt.ylim(-10, 10)



# Pearson
print(f"Pearson\ncorr_pearson = {corr}\np_value_pearson = {p_value}")

# Spearman
corrs, p_values = spearmanr(X, Y)
print(f"\nSpearmon\ncorr_spearman = {corrs}\np_value_spearman = {p_values}")

string = 'Pearson_corr = '+ str(corr)
plt.text(-0.2,-5, string, fontsize=15)

string = 'Spearman_corr = '+ "{:.2f}".format(corrs)
plt.text(-0.2,-6.5, string, fontsize=15)


plt.show(True)
````
